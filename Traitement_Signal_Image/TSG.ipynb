```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, ifft, rfftfreq
from numpy.fft import *
from scipy import signal
from sympy import *
import scipy.integrate as integrate
```

```
def xy(t,T):
  x = []
  y = []
  for i in t:
    if i>=-T and i<=T:
      y.append(1)
    else:
      y.append(0)

    if i>=-(T/2) and i<=(T/2):
      x.append(1)
    else:
      x.append(0)
  return np.array(x),np.array(y)       
```

```
T = 4
t = np.arange(-5,5,0.01)
x,y = xy(t,T)
```

```
plt.plot(x)
plt.xlabel("t")
plt.ylabel("X(t)")
plt.show()
```

```
plt.plot(y)
plt.xlabel("t")
plt.ylabel("X(t)")
plt.show()
```

```
corr_xy = signal.correlate(x,y)
corr_yx = signal.correlate(y,x)
conv = signal.convolve(x,y)
print(corr_xy.shape,corr_yx.shape,conv.shape)
```

```
plt.plot(np.linspace(-10,10,1999),conv)
plt.xlabel("t")
plt.ylabel("Convolution")
plt.show()
```

```
plt.plot(np.linspace(-10,10,1999),corr_xy)
plt.plot(np.linspace(-10,10,1999),corr_yx)
plt.xlabel("t")
plt.ylabel("Rxy/Ryx")
plt.show()
```

```
plt.plot(np.linspace(-5,5,1000),fft(x,1000))
plt.xlabel("t")
plt.ylabel("z")
plt.show()

plt.plot(np.linspace(-5,5,1000),fft(y,1000))
plt.xlabel("t")
plt.ylabel("z")
plt.show()
```
