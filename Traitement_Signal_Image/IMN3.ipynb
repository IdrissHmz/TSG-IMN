```python
f = np.fft.fft2 (pb_img)   # fournit la transformée de fréquence qui est un tableau complexe
fshift = np.fft.fftshift (f)  
# pour centrer on décale le résultat de N/2 dans les deux sens
module = 20*np.log (np.abs(fshift))
argument = np.angle(fshift)
```

```python
plt.figure(figsize=(10,10)) 
plt.subplot(1, 3 , 1)
plt.imshow( pb_img , cmap = 'gray')
plt.subplot(1,3,2)
plt.imshow(module, cmap = 'gray')
plt.subplot(1,3,3)
plt.imshow(argument, cmap = 'gray')
plt.show()
```
